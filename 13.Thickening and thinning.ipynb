```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow
def thickening(image, kernel_size):
  kernel = np.ones((kernel_size, kernel_size), np.uint8)
  thickened_image = cv2.dilate(image, kernel, iterations=1)
  return thickened_image
def thinning(image, kernel_size):
  kernel = np.ones((kernel_size, kernel_size), np.uint8)
  thinned_image = cv2.erode(image, kernel, iterations=1)
  return thinned_image
image = cv2.imread("/content/1.jpg", cv2.IMREAD_GRAYSCALE)
_, binary_image = cv2.threshold(image, 127, 255, cv2.THRESH_BINARY)
thickened_image = thickening(binary_image, kernel_size=3)
thinned_image = thinning(binary_image, kernel_size=3)
cv2_imshow(image)
cv2_imshow(thickened_image)
cv2_imshow(thinned_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
